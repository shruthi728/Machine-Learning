```python
from sklearn import datasets
import pandas as pd
```

```python
cancer=datasets.load_breast_cancer()
cancer
```

```python
bcancer=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
bcancer
```

```python
target=pd.DataFrame(cancer['target'])
target.columns=['target']
cancer=bcancer.join(target)
```

```python
cancer.groupby('target').count()
```

```python
x=cancer.drop(['target'],axis=1)
y=cancer['target']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_valid_test,y_train,y_valid_test=train_test_split(x,y,test_size=0.3)
```

```python
x_valid,x_test,y_valid,y_test=train_test_split(x_valid_test,y_valid_test,test_size=0.5)
```

```python
from sklearn.svm import SVC #SVC=support vector classifier
model=SVC()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_valid_test)
pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print('classification report')
print(classification_report(y_valid_test,pred))
print('confusion matrix')
print(confusion_matrix(y_valid_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
from sklearn.model_selection import GridSearchCV
params = {
 'max_depth': [2, 3, 5, 10, 20],
 'min_samples_leaf': [5, 10, 20, 50, 100],
 'criterion': ["gini", "entropy"]
}
grid_search = GridSearchCV(estimator=dt, param_grid=params, verbose=1, scoring = "accuracy")
grid_search.fit(x,y)
```

```python
score_df = pd.DataFrame(grid_search.cv_results_)
score_df.head()
```

```python
score_df.nlargest(5,"mean_test_score")
```

```python
grid_search.best_estimator_
```

```python
DecisionTreeClassifier(max_depth=2, min_samples_leaf=50, random_state=42)
```

```python
dt_best = grid_search.best_estimator_
from sklearn.metrics import classification_report
print(classification_report(y_test, dt_best.predict(x_test)))
```
